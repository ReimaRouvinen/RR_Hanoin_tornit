# HANOIN TORNIT

Hanoin tornit -pelin keksi ranskalainen matemaatikko Édouard Lucas vuonna 1883. Se perustuu vanhaan legendaan hindutemppelistä, jonka papeilla tai munkeilla on ratkaistavanaan 64 kultaisesta kiekosta koostuva Hanoin torni. Kun he saavat työnsä päätökseen, koittaa maailmanloppu. Ei tiedetä, keksikö Lucas tämän tarinan itse vai kehittikö hän pelin sen pohjalta. 

Lähde: <a href="https://fi.wikipedia.org/wiki/Hanoin_torni">Hanoin torni - Wikipedia</a>

```python
# Suorita kaikki ohjelmasolut valitsemalla valikosta Cell ja Run All

# tulosta tornit
def tulosta(torni1, torni2, torni3): 
    # tornien oikeat nimet
    text_1 = torni1[1]
    text_2 = torni2[1]
    text_3 = torni3[1]
    max = 0
    # tornien pituudet"
    p1 = len(torni1[0]) 
    p2 = len(torni2[0])
    p3 = len(torni3[0])  
    # selvitä korkeimman tornin korkeus
    if p1 > p2:
        if p1 > p3:
            max = p1
        else:
            max = p3
    else:
        if p2 > p3:
            max = p2
        else:
            max = p3        
    # tulosta tornit rivi kerrallaan
    for i in range(max):
        if p1 < max-i:
            if text_1 == "Torni1":
                text1 = "       "
            if text_1 == "Torni2":
                text2 = "       "
            if text_1 == "Torni3":
                text3 = "       " 
        else: 
            if text_1 == "Torni1":
                text1 = "{:^7}".format(torni1[0][max-1-i])
            if text_1 == "Torni2":            
                text2 = "{:^7}".format(torni1[0][max-1-i]) 
            if text_1 == "Torni3":            
                text3 = "{:^7}".format(torni1[0][max-1-i])  
        if p2 < max-i:
            if text_2 == "Torni1":
                text1 = "       "
            if text_2 == "Torni2":
                text2 = "       "
            if text_2 == "Torni3":
                text3 = "       "
        else: 
            if text_2 == "Torni1":
                text1 = "{:^7}".format(torni2[0][max-1-i]) 
            if text_2 == "Torni2":
                text2 = "{:^7}".format(torni2[0][max-1-i])  
            if text_2 == "Torni3":
                text3 = "{:^7}".format(torni2[0][max-1-i])
        if p3 < max-i:
            if text_3 == "Torni1":
                text1 = "       "
            if text_3 == "Torni2":
                text2 = "       "
            if text_3 == "Torni3":
                text3 = "       "
        else:  
            if text_3 == "Torni1":
                text1 = "{:^7}".format(torni3[0][max-1-i])
            if text_3 == "Torni2":
                text2 = "{:^7}".format(torni3[0][max-1-i]) 
            if text_3 == "Torni3":
                text3 = "{:^7}".format(torni3[0][max-1-i]) 
        print(text1+" "+text2+" "+text3) 
        
    print("TORNI1  TORNI2  TORNI3")
    print("")
    
print("Aliohjelma (tulosta) on määritelty")
```

```python
# tee siirrot
def hanoi(n, torni1, torni2, torni3, siirtoja, lkm):
    if n > 0:
        # siirrä n-1 kokoa oleva torni torni1-alustalta torni2-alustalle
        hanoi(n - 1, torni1, torni3, torni2, siirtoja, lkm)
        # siirrä kiekko torni1-alustalta torni3-alustalle
        if torni1[0]:
            kiekko = torni1[0].pop()
            siirtoja[0] = siirtoja[0] + 1
            siirrot = "Siirto nro {} - ".format(siirtoja[0])
            print(siirrot+"kiekko " + str(kiekko) + ": " + torni1[1] + " => " + torni3[1])
            print("")
            torni3[0].append(kiekko)
            if lkm < 16:
                # tulosta tornit
                tulosta(torni1, torni2, torni3)
        # siirrä n-1 kokoa oleva torni torni2-alustalta torni3-alustalle
        hanoi(n - 1, torni2, torni1, torni3, siirtoja, lkm) 
        
print("Aliohjelma (hanoi) on määritelty")
```

```python
# Voit muuttaa kiekkojen lukumäärää muuttamalla muuttujan lkm arvoa. 
# Suorita tämän jälkeen ohjelmasolu uudestaan valitsemalla Run.
# Huomaa, että kiekkojen määrän lisääminen lisää erittäin voimakkaasti siirtojen määrää
# ja yli 15 kiekolla tietokoneesi saattaa jumittua pahasti!
# Ohjelma tulostaa yli 15 kiekolla vain siirrot, ei torneja.

lkm = 3

# **************************************************************************************
siirtoja = [0]
# rakenna tornien alkutilanne
torni1 = ([], "Torni1")
for i in range(lkm):
    torni1[0].append(lkm-i)
torni2 = ([], "Torni2")    
torni3 = ([], "Torni3") 
# tulosta tornien alkutilanne
tulosta(torni1, torni2, torni3)
# tee siirrot
hanoi(len(torni1[0]),torni1,torni2,torni3, siirtoja, lkm)

print("Valmis")
```
