```python
# Suorita kaikki ohjelmasolut valitsemalla valikosta Cell ja Run All

def tulosta(source, helper, target): 
    
    text_s = source[1]
    text_h = helper[1]
    text_t = target[1]
    max = 0
    s = len(source[0]) 
    h = len(helper[0])
    t = len(target[0])  
    
    if s > h:
        if s > t:
            max = s
        else:
            max = t
    else:
        if h > t:
            max = h
        else:
            max = t        
    
    for i in range(max):
        if s < max-i:
            if text_s == "Lähde":
                text1 = "     "
            if text_s == "Apu":
                text2 = "     "
            if text_s == "Kohde":
                text3 = "     " 
        else: 
            if text_s == "Lähde":
                text1 = "{:^5}".format(source[0][max-1-i])
            if text_s == "Apu":            
                text2 = "{:^5}".format(source[0][max-1-i]) 
            if text_s == "Kohde":            
                text3 = "{:^5}".format(source[0][max-1-i])  
        if h < max-i:
            if text_h == "Lähde":
                text1 = "     "
            if text_h == "Apu":
                text2 = "     "
            if text_h == "Kohde":
                text3 = "     "
        else: 
            if text_h == "Lähde":
                text1 = "{:^5}".format(helper[0][max-1-i]) 
            if text_h == "Apu":
                text2 = "{:^5}".format(helper[0][max-1-i])  
            if text_h == "Kohde":
                text3 = "{:^5}".format(helper[0][max-1-i])
        if t < max-i:
            if text_t == "Lähde":
                text1 = "     "
            if text_t == "Apu":
                text2 = "     "
            if text_t == "Kohde":
                text3 = "     "
        else:  
            if text_t == "Lähde":
                text1 = "{:^5}".format(target[0][max-1-i])
            if text_t == "Apu":
                text2 = "{:^5}".format(target[0][max-1-i]) 
            if text_t == "Kohde":
                text3 = "{:^5}".format(target[0][max-1-i]) 
        print(text1+" "+text2+" "+text3) 
    print("Lähde  Apu  Kohde")
    print("")
    
print("Aliohjelma (tulosta) on määritelty")
```

```python
def hanoi(n, source, helper, target, siirtoja, lkm):
    if n > 0:
        # siirrä n-1 kokoa oleva torni Lähde-alustalta Apu-alustalle:
        hanoi(n - 1, source, target, helper, siirtoja, lkm)
        # siirrä kiekko Lähde-alustalta Kohde-alustalle
        if source[0]:
            disk = source[0].pop()
            siirtoja[0] = siirtoja[0] + 1
            siirrot = "Siirto nro {} - ".format(siirtoja[0])
            print(siirrot+"kiekko " + str(disk) + ": " + source[1] + " => " + target[1])
            print("")
            target[0].append(disk)
            if lkm < 16:
                tulosta(source, helper, target)
        # siirrä n-1 kokoa oleva torni Apu-alustalta Kohde-alustalle
        hanoi(n - 1, helper, source, target, siirtoja, lkm) 
        
print("Aliohjelma (hanoi) on määritelty")
```

```python
# Voit muuttaa kiekkojen lukumäärää muuttamalla muuttujan lkm arvoa. 
# Suorita tämän jälkeen ohjelmasolu uudestaan valitsemalla Run.
# Huomaa, että kiekkojen määrän lisääminen lisää erittäin voimakkaasti siirtojen määrää
# ja yli 15 kiekolla tietokoneesi saattaa jumittua pahasti!

lkm = 3

# *************************************************************************************************

siirtoja = [0]
source = ([], "Lähde")
for i in range(lkm):
    source[0].append(lkm-i)
helper = ([], "Apu")    
target = ([], "Kohde") 

tulosta(source, helper, target)

hanoi(len(source[0]),source,helper,target, siirtoja, lkm)

print("Valmis")
```
